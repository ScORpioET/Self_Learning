```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# 波士頓房價Dataset
data = load_boston()
clf = LinearRegression()
```

```python
x = data['data']
y = data['target']

clf.fit(x,y)
```

```python
y_pred = clf.predict(x[:20])
plt.figure(figsize=(10,5))
plt.plot(y_pred,linestyle = "--",color = 'g')
plt.plot(y[:20],color = 'r')

plt.show()

#虛線為預測
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
y_pred = clf.predict(x)
print('MSE',mean_squared_error(y_pred,y))# 均方誤差
print("MAE",mean_absolute_error(y_pred,y))# 絕對平均誤差
print("r2_score",r2_score(y_pred,y))# 決定係數，越靠近1效果越好
```
